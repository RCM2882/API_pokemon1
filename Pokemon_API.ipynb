```python
import requests



def fetch_pokemon_data(pokemon_name):
    url = f"https://pokeapi.co/api/v2/pokemon/{pokemon_name.lower()}"
    response = requests.get(url)
    if response.status_code == 200:
        return response.json()
    else:
        print(f"Failed to get data for {pokemon_name}")
        return None

def retrieve_pokemon_info(pokemon_data):
    name = pokemon_data['name']
    abilities = [ability['ability']['name'] for ability in pokemon_data['abilities']]
    types = [type_info['type']['name'] for type_info in pokemon_data['types']]
    weight = pokemon_data['weight']
    return name, abilities, types, weight

def clan_pokemon_by_type(pokemon_data):
    clans_of_pokemon = {}
    for pokemon in pokemon_data:
        name, abilities, types, weight = retrieve_pokemon_info(pokemon_data[pokemon])
        for pokemon_type in types:
            if pokemon_type not in clans_of_pokemon:
                clans_of_pokemon[pokemon_type] = {}
            clans_of_pokemon[pokemon_type][name] = {'abilities': abilities, 'weight': weight}
    return clans_of_pokemon

def create_pokemon(pokemon_names):
    pokemon_data = {}
    for name in pokemon_names:
        data = fetch_pokemon_data(name)
        if data:
            pokemon_data[name] = data
    return pokemon_data

def organize_pokemon_data(pokemon_data):
    clans_of_pokemon = clan_pokemon_by_type(pokemon_data)
    organized_data = {}
    for pokemon_type, pokemon_info in clans_of_pokemon.items():
        organized_data[pokemon_type.capitalize()] = pokemon_info
    return organized_data

def playtime():
    pokemon_names = ['bulbasaur', 'charmander', 'squirtle', 'pikachu', 'jigglypuff', 'meowth', 'psyduck', 'machop', 'bellsprout', 'tentacool', 'geodude', 'gastly', 'onix', 'drowzee', 'krabby', 'voltorb', 'exeggcute', 'cubone', 'koffing', 'rhyhorn']

    pokemon_data = create_pokemon(pokemon_names)
    organized_data = organize_pokemon_data(pokemon_data)

    for pokemon_type, pokemon_info in organized_data.items():
        print(f"{pokemon_type}:")
        for pokemon, info in pokemon_info.items():
            print(f"\t{pokemon}:")
            print(f"\t\tAbilities: {', '.join(info['abilities'])}")
            print(f"\t\tWeight: {info['weight']}")
            print()

if __name__ == '__main__':
    playtime()
```
